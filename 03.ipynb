```python
import numpy as np
from matplotlib import pyplot as plt

from filters import LowPassFilter 
from utils import load_sonar_data
```

```python
dt, ts, sonar_data = load_sonar_data()

ema_09 = LowPassFilter(0.9)
ema_04 = LowPassFilter(0.4)

filtered_09 = []
filtered_04 = []
for x in sonar_data:
    ema_09.update(x)
    ema_04.update(x)
    filtered_09.append(ema_09.get())
    filtered_04.append(ema_04.get())
    

fig = plt.figure(figsize=[16, 9])
plt.plot(ts, sonar_data, 'ro', label="measured")
plt.plot(ts, filtered_04, 'b-', label="alpha=0.4")
plt.plot(ts, filtered_09, 'g--', label="alpha=0.9")
plt.legend()
plt.ylabel("Distance[m]")
plt.xlabel("Time[sec]")
plt.show()
```
