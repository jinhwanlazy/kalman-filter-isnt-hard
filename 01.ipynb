# 예제 1-2

```python
dt = 0.2
ts = np.arange(0, 10, dt)
af = AverageFilter()
volts_averaged = []
volts_measured = []

for t, v in zip(ts, generate_volt_data()):
    af.update(v)
    
    volts_measured.append(v)
    volts_averaged.append(af.get())

fig = plt.figure(figsize=[12, 6])
plt.plot(ts, volts_measured, 'r*--', label="measured")
plt.plot(ts, volts_averaged, 'bo-', label="averaged")
plt.legend()
plt.ylabel("Volt[V]")
plt.xlabel("Time[sec]")
plt.show()
```
