```python
from matplotlib import pyplot as plt

from filters import MovingAverageFilter
from utils import load_sonar_data
```

# 예제 2-1

```python
dt, ts, sonar_data = load_sonar_data()

ma = MovingAverageFilter(10)
averaged = []
for x in sonar_data:
    ma.update(x)
    averaged.append(ma.get())

fig = plt.figure(figsize=[16, 9])
plt.plot(ts, sonar_data, 'ro', label="measured")
plt.plot(ts, averaged, 'b-', label="averaged")
plt.legend()
plt.ylabel("Distance[m]")
plt.xlabel("Time[sec]")
plt.show()
```
